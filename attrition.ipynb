## Part 1: Preprocessing

```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import numpy as np
from tensorflow.keras.models import Model
from tensorflow.keras import layers
from sklearn.model_selection import train_test_split

#  Import and read the attrition data
attrition_df = pd.read_csv('https://static.bc-edx.com/ai/ail-v-1-0/m19/lms/datasets/attrition.csv')
attrition_df.head()
```

```python
# Determine the number of unique values in each column.
attrition_df.nunique()
```

```python
# I used the Google Colab AI tool to help generate this code.
# prompt: # Create y_df with the Attrition and Department columns

y_df = attrition_df[['Attrition', 'Department']]
```

```python
# I used the Google Colab AI tool to help generate this code.
# prompt: # Create a list of at least 10 column names to use as X data
# # Create X_df using your selected columns
# # Show the data types for X_df

# Create a list of at least 10 column names to use as X data
selected_columns = ['Education', 'Age', 'DistanceFromHome', 'JobSatisfaction', 'OverTime', 'StockOptionLevel',
                   'WorkLifeBalance', 'YearsAtCompany', 'YearsSinceLastPromotion', 'NumCompaniesWorked']

# Create X_df using your selected columns
X_df = attrition_df[selected_columns]

# Show the data types for X_df
X_df.dtypes
```

```python
# I used the Google Colab AI tool to help generate this code.
# prompt:  # Convert your X data to numeric data types int64

#convert the OverTime column using one hot encoding, for Yes= 1 and No= 0 as an int64
X_df['OverTime'] = X_df['OverTime'].map({'Yes': 1, 'No': 0}).astype(int)

```

```python
# I used the Google Colab AI tool to help generate this code.
# prompt: # Create a StandardScaler
# # Fit the StandardScaler to the training data
# # Scale the training and testing data

# Split our preprocessed data into our features and target arrays
y = attrition_df[['Attrition','Department']]
X = X_df

# Split the preprocessed data into a training and testing dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)

# Create a StandardScaler instances
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
# I used the Google Colab AI tool to help generate this code.
# My instructor, Bill, helped me with this code, too.
# prompt: # Create a OneHotEncoder for the Department column
# # Fit the encoder to the training data
# # Create two new variables by applying the encoder
# # to the training and testing data

from sklearn.preprocessing import OneHotEncoder

# Create a OneHotEncoder instance
enc = OneHotEncoder(handle_unknown='ignore')

# Fit the encoder to the Department column in the training data
enc.fit(np.array(y_train['Department']).reshape(-1, 1))

# Create new variables by applying the encoder to the training and testing data
encoded_dept_train = enc.transform(np.array(y_train['Department']).reshape(-1, 1)).toarray()
encoded_dept_test = enc.transform(np.array(y_test['Department']).reshape(-1, 1)).toarray()

encoded_dept_train
```

```python
# Create a OneHotEncoder for the Attrition column
from sklearn.preprocessing import OneHotEncoder

# Create a OneHotEncoder instance
enc = OneHotEncoder(handle_unknown='ignore')

# Fit the encoder to the Department column in the training data
enc.fit(np.array(y_train['Attrition']).reshape(-1, 1))

# Create new variables by applying the encoder to the training and testing data
encoded_attrit_train = enc.transform(np.array(y_train['Attrition']).reshape(-1, 1)).toarray()
encoded_attrit_test = enc.transform(np.array(y_test['Attrition']).reshape(-1, 1)).toarray()

encoded_attrit_train

```

## Create, Compile, and Train the Model

```python
# I used the Google Colab AI tool to help generate this code.
# prompt: # Find the number of columns in the X training data
number_input_features = len(X_train.columns)
number_input_features
```

```python
# I used the Google Colab AI tool to help generate this code.
# prompt: # Create the input layer
input_layer = layers.Input(shape=(number_input_features,))

# Create at least two shared layers
shared_layer_1 = layers.Dense(units=64, activation='relu')(input_layer)
shared_layer_2 = layers.Dense(units=128, activation='relu')(shared_layer_1)

# Create separate output layers
output_layer = layers.Dense(units=1, activation='sigmoid')(shared_layer_2)

```

```python
# I used the Google Colab AI tool to help generate this code.
# I worked with my instructor, Bill, on this code, too
# prompt: # Create a branch for Department
# # with a hidden layer and an output layer
# # Create the hidden layer
# # Create the output layer

# Create the hidden layer for the Department branch
department_hidden = layers.Dense(units=32, activation='relu')(shared_layer_2)

# Create the output layer for the Department branch
department_output = layers.Dense(units=len(encoded_dept_train[1]), activation='sigmoid', name='department_output')(department_hidden)
```

```python
# I used the Google Colab AI tool to help generate this code.
# I worked with my instructor, Bill, on this code, too.
# prompt: # Create a branch for Attrition
# # with a hidden layer and an output layer
# # Create the hidden layer
# # Create the output layer

# Create the hidden layer for the Attrition branch
attrition_hidden = layers.Dense(units=32, activation='relu')(shared_layer_2)

# Create the output layer for the Attrition branch
attrition_output = layers.Dense(units=len(encoded_attrit_train[1]), activation='sigmoid', name='attrition_output')(attrition_hidden)
```

```python
# I used the Google Colab AI tool to help generate this code.
# prompt: # Create the model
# # Compile the model
# # Summarize the model

# Create the model
model = Model(inputs=input_layer, outputs=[department_output, attrition_output])

# Compile the model
model.compile(optimizer='adam',
              metrics={'department_output':'accuracy','attrition_output': 'accuracy'},
              loss={'department_output':'categorical_crossentropy','attrition_output': 'binary_crossentropy'}
              )

# Print a summary of the model
model.summary()

```

```python
# I used the Google Colab AI tool to help generate this code.
# prompt: # Train the model using the preprocessed data

fit_model = model.fit(X_train_scaled, {'department_output': encoded_dept_train, 'attrition_output' : encoded_attrit_train}, epochs=100)
```

```python
# Evaluate the model with the testing data
# include loss, department_output_loss, attrition_output_loss, department_output_accuracy and attrition_output_accuracy
# model_loss, department_output_loss, attrition_output_loss, department_output_accuracy, attrition_output_accuracy = model.evaluate(X_test_scaled, [encoded_dept_test, encoded_attrit_test])
model_details = model.evaluate(X_test_scaled, {'department_output': encoded_dept_test,'attrition_output': encoded_attrit_test})
# model_loss, model_accuracy, model_accuracy = model.evaluate(X_test_scaled, [encoded_dept_test, encoded_attrit_test])


model_details
```

```python
# Print the accuracy for both department and attrition
print(f"Department Accuracy: {model_details[2]}")
print(f"Attrition Accuracy: {model_details[1]}")

```
